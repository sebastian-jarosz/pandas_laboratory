```python
# Author - Sebastian Jarosz
# Student Nr - 78797

import numpy as np
import pandas as pd


# Zadanie nr 3 - START

# Part 1.	Wybrać trzy odrębne zbiory danych.
df_one = pd.read_csv('https://raw.githubusercontent.com/vaksakalli/datasets/master/airline_passenger_satisfaction.csv')
df_two = pd.read_csv('https://raw.githubusercontent.com/vaksakalli/datasets/master/melbourne_house_prices.csv')
df_three = pd.read_csv('https://raw.githubusercontent.com/vaksakalli/datasets/master/speed_dating.csv')

# Part 2.	Dla każdego zbioru danych podać wartości statystyczne dla wybranych 3 kolumn (jeżeli takie występują).

print('Ponizsze wartości mogą równiez zostać wydobyte za pomocą funkcji takich jak max(), min(), mean() itp.')

print('--------------------------------')
# airline_passenger_satisfaction.csv
print('Zbiór airline_passenger_satisfaction.csv')
print("Kolumna 'age'")
print(df_one['age'].describe())
print("Kolumna 'flight_distance'")
print(df_one['flight_distance'].describe())
print("Kolumna 'departure_delay_in_minutes'")
print(df_one['departure_delay_in_minutes'].describe())

print('--------------------------------')
# melbourne_house_prices.csv
print('Zbiór melbourne_house_prices.csv')
print("Kolumna 'BuildingArea'")
print(df_two['BuildingArea'].describe())
print("Kolumna 'Distance'")
print(df_two['Distance'].describe())
print("Kolumna 'Price'")
print(df_two['Price'].describe())

print('--------------------------------')
# speed_dating.csv
print('Zbiór speed_dating.csv')
print("Kolumna 'age'")
print(df_three['age'].describe())
print("Kolumna 'age_o'")
print(df_three['age_o'].describe())
print("Kolumna 'd_age'")
print(df_three['d_age'].describe())

# Part 3.	Dla każdego zbioru wybrać co najmniej 3 kolumny, dla których przeprowadzone zostanie grupowanie. Dla każdej z grup wykonać:
# a)	Wypisać wartość minimalną pierwszego kwartyla dla wybranych dwóch kolumn spoza grupy.
# b)	Wypisać wartość maksymalną drugiego kwartyla dla wybranych dwóch kolumn spoza grupy.
# c)	Wypisać wartość średnią trzeciego kwartyla dla wybranych dwóch kolumn spoza grupy.
# d)    Wypisać wariancję z czwartego kwartyla dla wybranych dwóch kolumn spoza grupy

print('--------------------------------')
# airline_passenger_satisfaction.csv
print('Zbiór airline_passenger_satisfaction.csv')
print("Grupowanie po 'Gender', 'type_of_travel', 'customer_class' - Kolumny 'flight_distance', 'age'")
group_one = df_one.groupby(['Gender', 'type_of_travel', 'customer_class'])[['flight_distance', 'age']]

print("---Pierwszy kwartyl---")
group_one_q1 = group_one.agg(lambda x: np.quantile(x, q=0.25))
# print(group_one_q1)
print("Wartość minimalna pierwszego kwartyla dla kolumn 'flight_distance', 'age':")
print(group_one_q1.min())

print("---Drugi kwartyl---")
group_one_q2 = group_one.agg(lambda x: np.quantile(x, q=0.50))
# print(group_one_q2)
print("Wartość maksymalna drugiego kwartyla dla kolumn 'flight_distance', 'age':")
print(group_one_q2.max())

print("---Trzeci kwartyl---")
group_one_q3 = group_one.agg(lambda x: np.quantile(x, q=0.75))
# print(group_one_q3)
print("Wartość średnia trzeciego kwartyla dla kolumn 'flight_distance', 'age':")
print(group_one_q3.mean())

print("---Czwarty kwartyl---")
group_one_q4 = group_one.agg(lambda x: np.quantile(x, q=1))
# print(group_one_q4)
print("Wariancja czwartego kwartyla dla kolumn 'flight_distance', 'age':")
print(group_one_q4.var())

print('--------------------------------')
# melbourne_house_prices.csv
print('Zbiór melbourne_house_prices.csv')
print("Grupowanie po 'Suburb', 'SellerG', 'Regionname' - Kolumny 'BuildingArea', 'Price'")
group_two = df_two.groupby(['Suburb', 'SellerG', 'Regionname'])[['BuildingArea', 'Price']]

print("---Pierwszy kwartyl---")
group_two_q1 = group_two.agg(lambda x: np.quantile(x, q=0.25))
# print(group_two_q1)
print("Wartość minimalna pierwszego kwartyla dla kolumn 'BuildingArea', 'Price':")
print(group_two_q1.min())

print("---Drugi kwartyl---")
group_two_q2 = group_two.agg(lambda x: np.quantile(x, q=0.50))
# print(group_two_q2)
print("Wartość maksymalna drugiego kwartyla dla kolumn 'BuildingArea', 'Price':")
print(group_two_q2.max())

print("---Trzeci kwartyl---")
group_two_q3 = group_two.agg(lambda x: np.quantile(x, q=0.75))
# print(group_two_q3)
print("Wartość średnia trzeciego kwartyla dla kolumn 'BuildingArea', 'Price':")
print(group_two_q3.mean())

print("---Czwarty kwartyl---")
group_two_q4 = group_two.agg(lambda x: np.quantile(x, q=1))
# print(group_two_q4)
print("Wariancja czwartego kwartyla dla kolumn 'BuildingArea', 'Price':")
print(group_two_q4.var())


print('--------------------------------')
# speed_dating.csv
print('Zbiór speed_dating.csv')
print("Grupowanie po 'gender', 'race', 'field' - Kolumny 'pref_o_attractive', 'pref_o_intelligence'")
group_three = df_three.groupby(['gender', 'race', 'field'])[['pref_o_attractive', 'pref_o_intelligence']]

print("---Pierwszy kwartyl---")
group_three_q1 = group_three.agg(lambda x: np.quantile(x, q=0.25))
# print(group_three_q1)
print("Wartość minimalna pierwszego kwartyla dla kolumn 'pref_o_attractive', 'pref_o_intelligence':")
print(group_three_q1.min())

print("---Drugi kwartyl---")
group_three_q2 = group_three.agg(lambda x: np.quantile(x, q=0.50))
# print(group_three_q2)
print("Wartość maksymalna drugiego kwartyla dla kolumn 'pref_o_attractive', 'pref_o_intelligence':")
print(group_three_q2.max())

print("---Trzeci kwartyl---")
group_three_q3 = group_three.agg(lambda x: np.quantile(x, q=0.75))
# print(group_three_q3)
print("Wartość średnia trzeciego kwartyla dla kolumn 'pref_o_attractive', 'pref_o_intelligence':")
print(group_three_q3.mean())

print("---Czwarty kwartyl---")
group_three_q4 = group_three.agg(lambda x: np.quantile(x, q=1))
# print(group_three_q4)
print("Wariancja czwartego kwartyla dla kolumn 'pref_o_attractive', 'pref_o_intelligence':")
print(group_three_q4.var())

# Part 5. Dla każdego zbioru wybrać dwie kolumny, dla których przeprowadzone zostanie grupowanie. Wypisać wszystkie wartości wewnątrz grupy oraz ilość elementów wewnątrz grupy.

print('--------------------------------')
# airline_passenger_satisfaction.csv
print('Zbiór airline_passenger_satisfaction.csv')
print("Grupowanie po 'Gender', 'type_of_travel'")
print(df_one.groupby(['Gender', 'type_of_travel']).count())


print('--------------------------------')
# melbourne_house_prices.csv
print('Zbiór melbourne_house_prices.csv')
print("Grupowanie po 'Suburb', 'Regionname'")
print(df_two.groupby(['Suburb', 'Regionname']).count())

print('--------------------------------')
# speed_dating.csv
print('Zbiór speed_dating.csv')
print("Grupowanie po 'gender', 'field'")
print(df_three.groupby(['gender', 'field']).count())
```
